```python
#Import necessary Libraries
import pandas as pd
```

```python
# Load dataset 
df1 = pd.read_csv("vgsales.csv")
```

```python
# display top 5 rows
df1.head()
```

```python
# Creating a new column "Period" where year<=2005 is labelled as 'Pre_2005' and year > 2005 labelled as 'Post_2005'
df1['Period'] = df1['Year'].apply(lambda x: 'Pre_2005' if x <= 2005 else 'Post_2005')
#Displaying top 5 rows of df1
df1.head()
```

```python
# calculating average global sales for Pre_2005 and storing the result in avg_sales_Pre_2005
avg_sales_Pre_2005 = df1.loc[df1['Period'] == 'Pre_2005', 'Global_Sales'].mean()
# calculating average global sales for Pre_2005 and storing the result in avg_sales_Pre_2005
avg_sales_Post_2005 = df1.loc[df1['Period'] == 'Post_2005', 'Global_Sales'].mean()
# printing avg_sales_Pre_2005
print(avg_sales_Pre_2005)
# printing avg_sales_Post_2005
print(avg_sales_Post_2005)
```

```python
# Comparing the averaege global sales using if and else statements and printing the message output
if avg_sales_Pre_2005 < avg_sales_Post_2005:
    print(" The Average Global Sales are higher After 2005")
else:
    print(" The Average Global Sales are higher Before 2005")
```

```python
# query to create a new column "Time_Period" using CASE and WHEN
# Calculate the average using avg()
df = pd.read_sql_query("""
    SELECT 
        CASE 
            WHEN Year <= 2005 THEN 'Pre_2005'  
            ELSE 'Post_2005'
        END AS Time_Period,
        AVG(Global_Sales) AS Avg_Global_Sales
    FROM vgsales
    GROUP BY Time_Period
""", conn)
```
